**4.2. Detecci칩n y tratamiento de valores at칤picos (outliers)**

**Variables num칠ricas**

```python
#Importar librer칤as
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Asegurarse de que los nombres de columnas est칠n limpios
df.columns = df.columns.str.strip()

# Variables num칠ricas a analizar
numericas = ['IQ','Prev_Sem_Result','CGPA','Academic_Performance','Extra_Curricular_Score','Communication_Skills','Projects_Completed'
]

# Funci칩n para detectar e imputar valores at칤picos usando la mediana
# Cuando se tienen valores at칤picos, es aconsejable imputar por la mediana
# ya que, los valores at칤picos afectan a la media

#def significa que se crear치 una funci칩n definida por el usuario
#Creamos una funci칩n llamada imputar_outliers_con_mediana y le pasamos como argumentos
#el DataFrame (df) y las columnas (col)
def imputar_outliers_con_mediana(df, col):
    df_copy = df.copy() #Hacemos una copia del dataset

    # Calcular l칤mites inferior y superior para obtener el rango intercuart칤lico (IQR)
    Q1 = df[col].quantile(0.25) # C치lculo del cuartil 1
    Q3 = df[col].quantile(0.75) # C치lculo del cuartil 2
    IQR = Q3 - Q1 # C치lculo del rango intercuart칤lico
    lim_inf = Q1 - 1.5 * IQR # L칤mite inferior - valores por debajo se consideran at칤picos
    lim_sup = Q3 + 1.5 * IQR # L칤mite superior -  valores por encima se consideran at칤picos

    # Identificar outliers
    outliers = df[(df[col] < lim_inf) | (df[col] > lim_sup)]
    print(f"{col}: {len(outliers)} valores at칤picos detectados.")

    # Calcular mediana sin outliers
    mediana = df[(df[col] >= lim_inf) & (df[col] <= lim_sup)][col].median()

    # Imputar con la mediana
    df_copy[col] = df_copy[col].apply(lambda x: mediana if x < lim_inf or x > lim_sup else x)
    print(f"Valores at칤picos de '{col}' imputados con la mediana: {mediana}\n")

    return df_copy

# Aplicar la funci칩n a cada variable num칠rica
for col in numericas: # Recorremos la lista de variables num칠ricas usando for
    df = imputar_outliers_con_mediana(df, col)

# Boxplots actualizados
for col in numericas:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot de {col} (sin valores at칤picos)')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

En los gr치ficos de cajas, se puede observar que no hay valores atipicos. As칤 que no hay necesidad de realizar imputaciones.

```python
# Tratar columnas categ칩ricas de manera individual
df['College_ID'] = df['College_ID'].astype(str).str.strip()

# Para columnas binarias, estandarizar valores
for col in ['Internship_Experience', 'Placement']:
    df[col] = df[col].astype(str).str.strip().str.lower()
    df[col] = df[col].replace({'yes': 'Yes', 'no': 'No', 's칤': 'Yes', 'si': 'Yes'})
    df[col] = df[col].astype('category')
    df
```

**5.1.2. Bivariado (Relacionar variables)**

Luego de examinar el comportamiento individual de cada variable mediante el an치lisis descriptivo univariado, el siguiente paso consiste en explorar c칩mo se relacionan entre s칤. Esta fase, es clave para identificar patrones, dependencias o asociaciones relevantes que puedan influir en el fen칩meno estudiado.

쮺칩mo se relacionan las variables?

La relaci칩n entre variables debe establecerse en funci칩n del objetivo del an치lisis o del problema que se desea resolver. Es fundamental seleccionar y analizar aquellas combinaciones de variables que puedan aportar informaci칩n significativa y 칰til para la comprensi칩n del contexto, la formulaci칩n de hip칩tesis o el desarrollo de modelos predictivos. Esta etapa permite detectar conexiones relevantes que pueden no ser evidentes al observar cada variable por separado.

En el data set bajo estudio, podr칤a ser interesante relacionar las siguientes variables, para observar si existen patrones.

image.png

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from itertools import combinations

# Cargar dataset (assuming df is not already loaded)
try:
    df
except NameError:
    df = pd.read_csv("/content/college_student_placement_datos.csv")


# Define a list of columns to analyze for relationships, excluding 'College_ID' for numerical plots
columns_to_analyze = ['IQ', 'Prev_Sem_Result', 'CGPA', 'Academic_Performance',
                      'Internship_Experience', 'Extra_Curricular_Score',
                      'Communication_Skills', 'Projects_Completed', 'Placement']

numerical_cols_for_regplot = ['IQ', 'Prev_Sem_Result', 'CGPA', 'Academic_Performance',
                              'Extra_Curricular_Score', 'Communication_Skills',
                              'Projects_Completed']

# Generate combinations of columns for bivariate analysis
combinations_list = list(combinations(columns_to_analyze, 2))

# Create heatmaps for each combination
for combo in combinations_list:
    plt.figure(figsize=(10, 6))
    # For categorical vs categorical or numerical vs categorical, use crosstab
    if df[combo[0]].dtype.name == 'category' or df[combo[1]].dtype.name == 'category':
        cross_tab = pd.crosstab(df[combo[0]], df[combo[1]])
        sns.heatmap(cross_tab, annot=True, fmt='d', cmap='YlGnBu')
    # For numerical vs numerical, use a scatter plot with regression line or a correlation heatmap
    elif combo[0] in numerical_cols_for_regplot and combo[1] in numerical_cols_for_regplot:
        sns.scatterplot(data=df, x=combo[0], y=combo[1])
        sns.regplot(data=df, x=combo[0], y=combo[1], scatter=False, color='red') # Add regression line
    else: # Handle other combinations like categorical vs numerical which are not handled by crosstab or regplot
        sns.scatterplot(data=df, x=combo[0], y=combo[1])

    plt.title(f'Relaci칩n entre {combo[0]} y {combo[1]}')
    plt.show()
```

A continuaci칩n, se presentan algunos ejemplos, combinando diferentes tipos de variables:
